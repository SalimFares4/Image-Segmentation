```python
from osgeo import gdal
import numpy as np
import rasterio

dem_path = '/content/drive/My Drive/Image Segmentation/'

aspect_path = dem_path + "aspect.tif"
slope_path = dem_path + "slope.tif"
file_path = dem_path + "Firsoff_HiRISE_5.tif-adj.tif"

def calculate_slope(DEM):
    gdal.DEMProcessing(slope_path, DEM, 'slope')
    with rasterio.open(slope_path) as dataset:
        slope=dataset.read(1)
    return slope

def calculate_aspect(DEM):
    gdal.DEMProcessing(aspect_path, DEM, 'aspect')
    with rasterio.open(aspect_path) as dataset:
        aspect=dataset.read(1)
    return aspect
slope=calculate_slope(file_path)
aspect=calculate_aspect(file_path)

print(type(slope))
print(slope.dtype)
print(slope.shape)
```

```python
import rasterio.plot

with rasterio.open(file_path) as dem:
  rasterio.plot.show(dem)

with rasterio.open(aspect_path) as aspect:
  rasterio.plot.show(aspect)

with rasterio.open(slope_path) as slope:
  rasterio.plot.show(slope)
```
